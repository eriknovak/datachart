# Color Themes

This notebook contains examples of the different color themes available in the `datachart` package.

Let's start by importing the necessary modules.

```python
import random
import numpy as np
from datachart import line_chart, bar_chart, histogram, heatmap
from datachart.constants import (
    VALFMT,
    FIG_SIZE,
    LINE_MARKER,
    LINE_STYLE,
    THEME
)
```

```python
from datachart.config import config
```

## Default Theme

The default theme is the theme, that the user will by default see when using the `datachart` package. No additional configuration is required.

In case if the user uses a different theme, or if they change it by hand, the default theme can be automatically reverted using the `config` module.


```python
# reset the config to the default theme
config.reset_config()
```

```python
# explicity set the default theme
config.set_theme(THEME.DEFAULT)
```

### Default Theme: Line Chart

```python
_ = line_chart(
    {
        "charts": [
            {
                "data": [{"x": x / 10, "y": np.cos(x / 2)} for x in range(21)],
                "style": {
                    "plot.line.marker": LINE_MARKER.POINT,
                    "plot.line.style": LINE_STYLE.DOTTED,
                },
                "subtitle": "cosine",  # add a subtitle to the line
            },
            {
                "data": [{"x": x / 10, "y": np.sin(x / 2)} for x in range(21)],
                "subtitle": "sine",  # add a subtitle to the line
            },
        ],
        "title": "Title",
        "xlabel": "the global x-axis label",
        "ylabel": "the global y-axis label",
        "figsize": FIG_SIZE.A4_NARROW,
        "show_grid": "both",
        "show_legend": True,  # show the legend
    }
)
```

### Default Theme: Bar Chart

```python
_ = bar_chart(
    {
        "charts": {
            "data": [
                {"label": f"xx{id}", "y": 100 * (id + 1) * random.random()}
                for id in range(10)
            ],
            "vlines": [{"x": 2 * i} for i in range(1, 4)],
            "hlines": {"y": 400},
        },
        "title": "Title",
        "xlabel": "the global x-axis label",
        "ylabel": "the global y-axis label",
        "figsize": FIG_SIZE.A4_NARROW,
        "show_grid": "both",
    }
)
```

### Default Theme: Histogram

```python
_ = histogram(
    {
        "charts": [
            {
                "data": {"x": [100 * idx * random.random() for idx in range(100)]},
                "subtitle": f"Histogram {id}",
            }
            for id in range(2)
        ],
        "title": "Title",
        "xlabel": "the global x-axis label",
        "ylabel": "the global y-axis label",
        "figsize": FIG_SIZE.A4_NARROW,  # must be a tuple (width, height), values are in inches
        "show_grid": "y",  # options: "both", "x", "y"
        "num_bins": 40,
    }
)
```

### Default Theme: Heatmap

```python
_ = heatmap(
    {
        "charts": {
            "data": [[x + y for x in range(8)] for y in range(8)],
            "valfmt": VALFMT.DECIMAL,
            "xticks": [idx for idx in range(8)],
            "xticklabels": [f"xxx{idx}" for idx in range(8)],
            "xtickrotate": 45,
            "yticks": [idx for idx in range(8)],
            "yticklabels": [f"xxx{idx}" for idx in range(8)],
            "ytickrotate": 0,
        },
        "title": "Title",
        "xlabel": "the global x-axis label",
        "ylabel": "the global y-axis label",
        "figsize": FIG_SIZE.SQUARE_SMALL,
        "show_heatmap_vals": True,
        "show_colorbars": True,
    }
)
```

## Colormaps

This section shows the different colormaps that are available in `datachart`. All colormaps are available in the `COLORS` constant.

```python
from datachart.constants import COLORS
```

```python
import numpy as np
import matplotlib.pyplot as plt
from datachart.utils.colors import get_colormap
```

```python
def plot_color_gradients(cmap_list, n_sections=256):

    gradient = np.linspace(0, 1, n_sections)
    gradient = np.vstack((gradient, gradient))

    # Create figure and adjust figure height to number of colormaps
    nrows = len(cmap_list)
    figh = 0.35 + 0.15 + (nrows + (nrows - 1) * 0.1) * 0.22
    fig, axs = plt.subplots(nrows=nrows + 1, figsize=(6.4, figh))
    fig.subplots_adjust(top=1 - 0.35 / figh, bottom=0.15 / figh,
                        left=0.2, right=0.99)
    axs[0].set_title(f'datachart colormaps', fontsize=14)

    for ax, name in zip(axs, cmap_list):
        ax.imshow(gradient, aspect='auto', cmap=get_colormap(name))
        ax.text(-0.01, 0.5, name, va='center', ha='right', fontsize=10,
                transform=ax.transAxes)

    # Turn off *all* ticks & spines, not just the ones with colormaps.
    for ax in axs:
        ax.set_axis_off()
```

```python
cmap_list = [
    COLORS.Blue,
    COLORS.Green,
    COLORS.Orange,
    COLORS.Purple,
    COLORS.Grey,
    COLORS.YlGnBu,
    COLORS.YlGn,
    COLORS.BuGn,
    COLORS.GnBu,
    COLORS.PuBu,
    COLORS.RdBn,
    COLORS.RdYlBu,
    COLORS.BrNg,
    COLORS.PrGn,
    COLORS.PuOr,
    COLORS.RdGy,
    COLORS.RdYlGn,
    COLORS.Spectral,
    COLORS.MixedLight,
    COLORS.MixedDark,
]
```

### Continuous scales

```python
plot_color_gradients(cmap_list=cmap_list)
```

### Discrete scales

```python
# change the number of sections `n_sections`
plot_color_gradients(cmap_list=cmap_list, n_sections=10)
```
